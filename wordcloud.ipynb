```python
with open(r'被讨厌的勇气.txt','r', encoding='utf-8') as f:
    txt = f.read()
print(len(txt))
```

```python
sentence = 'This article is very useful'
words = sentence.split(' ')
```

```python
# 获得停止词
with open('stopwords.txt', 'r', encoding='utf-8') as f:
    stopwords = f.readlines()
# 去掉多余的空格
for i in range(len(stopwords)):
    stopwords[i] = stopwords[i].strip()
```

```python
len(stopwords)
```

```python
import jieba
words_list = jieba.lcut(txt)
```

```python
import wordcloud
font = r'C:\Windows\Fonts\simhei.ttf'
wc = wordcloud.WordCloud(background_color='white',
                         scale=15,
                         font_path=font,
                         stopwords=stopwords,
                         max_words=100)
string = " ".join(words_list)
wc.generate(string)
# 保存成文件
wc.to_file('success.png')
```

```python
# 背景图
import imageio
image = imageio.imread('heart.jpg')

import wordcloud
font = r'C:\Windows\Fonts\simhei.ttf'
wc = wordcloud.WordCloud(background_color='white',
                         scale=15,
                         font_path=font,
                         stopwords=stopwords,
                         max_words=100,
                         mask=image,
                         colormap='Reds')
string = " ".join(words_list)
wc.generate(string)
# 保存成文件
wc.to_file('success2.png')
```

```python
import matplotlib.pyplot as plt 
fig, ax = plt.subplots(figsize=(10,12))
ax.imshow(wc.to_array())
ax.set_axis_off()
```
